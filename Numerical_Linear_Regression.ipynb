```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

pf=pd.read_csv('PIzza.csv')
x1=pf['Size'].mean()
x2=pf['Price'].mean()
np_size=np.array(pf.Size)
np_price=np.array(pf.Price)
mul=np_size*np_price
sqr=np_size*np_size
x3=np.average(mul)  #Claculate Average Using Numpy
x4=np.average(sqr)  #Calculate Average Using Numpy
x1_sqr=x1*x1
m=(x1*x2-x3)/(x1_sqr-x4)
c=x2-m*x1
m="%.2f" % m #Getting a Specific Digit After Decimal in floating point
c="%.1f" % c

plt.plot(pf.Size,pf.Price)
plt.xlabel('Pizzas Size')
plt.ylabel('Price')
plt.title('The Rate of Pizza with Respect ot its Size')
plt.show()
```
